# About Dataset
## Context
Kumpulan data berisi survei kepuasan penumpang maskapai penerbangan. Faktor apa yang sangat berkorelasi dengan kepuasan (atau ketidakpuasan) penumpang? Tujuan dari dataset untuk memprediksi kepuasan penumpang?

https://www.kaggle.com/datasets/teejmahal20/airline-passenger-satisfaction

## Content
`Gender`: Jenis kelamin penumpang (Female, Male)<br>
`Customer Type`: Tipe pelanggan (Loyal customer, disloyal customer)<br>
`Age`: Usia sebenarnya dari penumpang<br>
`Type of Travel`: Tipe penerbangan penumpang (Personal Travel, Business Travel)<br>
`Class`: Kelas perjalanan di pesawat penumpang (Business, Eco, Eco Plus)<br>
`Flight distance`: Jarak penerbangan dari perjalanan<br>
`Inflight wifi service`: Tingkat kepuasan layanan wifi dalam pesawat (0:Not Applicable;1-5)<br>
`Departure/Arrival time convenient`: Tingkat kepuasan waktu Keberangkatan/Kedatangan nyaman<br>
`Ease of Online booking`: Tingkat kepuasan pemesanan online<br>
`Gate location`: Tingkat kepuasan lokasi Gerbang<br>
`Food and drink`: Tingkat kepuasan Makanan dan minuman<br>
`Online boarding`: Tingkat kepuasan boarding online<br>
`Seat comfort`: Tingkat kepuasan kenyamanan tempat duduk<br>
`Inflight entertainment`: Tingkat kepuasan hiburan dalam pesawat<br>
`On-board service`: Tingkat kepuasan layanan On-board<br>
`Leg room service`: Tingkat kepuasan pelayanan kamar Kaki<br>
`Baggage handling`: Tingkat kepuasan penanganan bagasi<br>
`Check-in service`: Tingkat kepuasan pelayanan Check-in<br>
`Inflight service`: Tingkat kepuasan pelayanan dalam pesawat<br>
`Cleanliness`: Tingkat kepuasan Kebersihan<br>
`Departure Delay in Minutes`: Menit tertunda saat keberangkatan<br>
`Arrival Delay in Minutes`: Menit tertunda saat Kedatangan<br>
`Satisfaction`: Tingkat kepuasan maskapai penerbangan(Satisfaction, dissatisfaction)<br>

# Import Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv("data/train.csv", index_col="id")
df2 = pd.read_csv("data/test.csv", index_col="id")

df = pd.concat([df1, df2])
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
df.head()
```

```python
df.sort_values('id', inplace=True)
```

# Handle Missing Value

```python
df.dropna(axis=0, inplace=True)
```

```python
df.isnull().sum()
```

# EDA

```python
df.replace({'neutral or dissatisfied': 0, 'satisfied': 1}, inplace=True)
df.info()
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

## Target

```python
print(df.satisfaction.value_counts())
sns.countplot(x=df.satisfaction, data=df);
```

Dari data target yang kita punya di ketahui bahwa kebanyakan penumpang tidak puas terhadap maskapai penerbangan.

## Age

```python
plt.figure(figsize=(7, 6))
sns.distplot(df.Age[df.satisfaction == 0], bins=[0, 10, 30, 50, 85], color='r', label='Tidak Puas')
sns.distplot(df.Age[df.satisfaction == 1], bins=[0, 10, 30, 50, 85], color='g', label='Puas')
plt.legend();
```

Jika dilihat dari grafiknya, kebanyakan orang yang berumur 0-30 itu merasa tidak puas terhadap maskapai penerbangan. Tapi dari usia 31-85 hampir semua merasa puas terhadap maskapai penerbangan.

## Fitur Class terhadap Label

```python
sns.set_style('darkgrid')

fig, axs = plt.subplots(ncols=3,figsize=(15, 6))

category = ["Eco", "Eco Plus",  "Business"]

for i, cat in enumerate(category):
    sns.countplot(x='satisfaction', data=df[df['Class'] == cat], ax=axs[i], hue='satisfaction', palette='Set2')
    axs[i].set_title(cat)
    axs[i].legend(['Tidak Puas', 'Puas'])
    

plt.show()
```

Class yang digunakan dalam penerbangan sangat mempengaruhi tingkat kepuasan dari customer, Class Eco dan Eco Pluss menjadi Class yang memiliki tingkat ketidak puasan yang lebih tinggi, sedangkan Class Business hampir 60% dari customer merasa puas.

Mungkin ada beberapa fasilitas tambahan yang di sediakan atau sebuah prioritas terhadap customer Class Business, sehingga kenyamanan yang di berikan sesuai dengan harga yang di bayar.

## Fitur Class terhadap Customer Type

```python
sns.set_style('darkgrid')

fig, axs = plt.subplots(ncols=3,figsize=(15, 6))

category = ["Eco", "Eco Plus",  "Business"]

for i, cat in enumerate(category):
    sns.countplot(x='Customer Type', data=df[df['Class'] == cat], ax=axs[i], palette='Set2', order=['disloyal Customer', 'Loyal Customer'])
    axs[i].set_title(cat)
    
plt.show()
```

```python
sns.set_style('darkgrid')
sns.countplot(data=df, x=df.Class, palette='Set2');
```

Dari ketiga Class kita tau bahwa ketiganya memiliki Loyal Customer lebih banyak daripada Disloyal Customer, karena memang data yang kita miliki kebanyakan yang melakukan review terhadap maskapai penerbangan merupakan Loyal Customer. Jadi tidak heran kalau ketiganya memiliki Loyal Customer yang lebih banyak.

Pengguna Business memang terlihat yang paling banyak, namun antara Eco dan Eco Plus ada jarak yang signifikan. Kita tau Eco plus adalah class yang lebih tinggi dari Eco, akan tetap pengguna dengan Class Eco jauh lebih banyak.

Mungkin dari segi kenyamanan yang di tawarkan tidak semewah Business dan harga yang di patok lebih mahal dari pada Eco membuat customer berfikir 2 kali. Jika dengan kenyamanan yang di tambahkan sedikit dan harga yang tawarkan terbilang lebih mahal, maka Customer akan tetap memilih class Eco untuk menghemat uang mereka.

## Fitur Class terhadap Flight Distance

```python
plt.figure(figsize=(7, 6))
bins = list(range(0, 5001, 500))
sns.distplot(df['Flight Distance'][df.Class == 'Eco'], bins=bins, color='r', label='Eco')
sns.distplot(df['Flight Distance'][df.Class == 'Eco Plus'], bins=bins, color='g', label='Eco Plus')
sns.distplot(df['Flight Distance'][df.Class == 'Business'], bins=bins, color='b', label='Business')
plt.legend()
plt.show()
```

Bahkan jarak tempuh sangat mempengaruhi pemilihan Class pada Customer, kalau kita lihat Class Eco dan Eco Pluss rata-rata digunakan hanya untuk jarak tempuh yang relatif dekat. Namun jika jaraknya sudah sangat jauh, biasanya Customer memilih untuk membeli tiket Business karena kenyamanan dalam perjalanan jauh cukup penting.

## Fitur Type of Travel terhadap Class

```python
sns.set_style('darkgrid')
fig, axs = plt.subplots(ncols=2,figsize=(15, 7))

category = ["Personal Travel", "Business travel"]

for i, cat in enumerate(category):
    sns.countplot(x='Class', data=df[df['Type of Travel'] == cat], ax=axs[i], palette='Set2', order=['Eco', 'Eco Plus', 'Business'])
    axs[i].set_title(cat)
    
plt.show()
```

- **Personal Travel** </br> Personal Travel sendiri merujuk pada tujuan perjalanan untuk pribadi atau liburan, kebanyakakan dari mereka yang Personal Travel hanya akan membeli Class Eco untuk mengirit biaya transportasi. Sedikit orang yang memilih Eco Plus apalagi memilih Class Business untuk perjalanan pribadi.
- **Business Travel** </br> Business Travel merujuk pada tujuan perjalanan kepentingan bisnis. Dan sudah pasti jelas pada Business Travel Class yang dipilih adalah Business, namun juga ada Business Travel yang memilih untuk Class Eco, dan sangat sedikit untuk Eco Plus.

## Arrival Delay and Departure Delay

```python
# ARRIVAL

plt.figure(figsize=(7, 6))
bins = list(range(0, 1585, 60))
sns.distplot(df['Arrival Delay in Minutes'][df.satisfaction == 0], bins=bins, color='r', label='Tidak Puas')
sns.distplot(df['Arrival Delay in Minutes'][df.satisfaction == 1], bins=bins, color='g', label='Puas')
plt.legend()
plt.show()
```

```python
# DEPATURE

plt.figure(figsize=(7, 6))
bins = list(range(0, 1600, 60))
sns.distplot(df['Departure Delay in Minutes'][df.satisfaction == 0], bins=bins, color='r', label='Tidak Puas')
sns.distplot(df['Departure Delay in Minutes'][df.satisfaction == 1], bins=bins, color='g', label='Puas')
plt.legend()
plt.show()
```

```python
print("Arrival Delay: ", df['Arrival Delay in Minutes'].mean())
print("Depature Delay: ", df['Departure Delay in Minutes'].mean())
```

kalau kita lihat keduanya rata-rata keterlambatan maskapai itu di sekitar 14-15 Menit, dan ada beberapa maskapai yang keterlambatannya mencapai 1600 dalam Menit. Jika dilihat grafik di atas bagian kanan maka akan ada garis merah, yang sudah pasti jelas bahwa keterlambatan yang begitu lama sangat membuat customer merasa tidak puas.

## Fitur Like Rating Value

Maksudnya kebanyakan fitur di dalam dataset berupa nilai rating, bahkan 60%-70% dataset kita berupa rating dari semua fasilitas maskapai penerbangan, walaupun saya pikir mungkin ada beberapa fasilitas yang tidak di masukan ke dalam dataset.

Tingkat Rating

- `0` Sangat Tidak Bagus
- `1` Tidak Bagus
- `2` Cukup Bagus
- `3` Bagus
- `4` Sangat Bagus
- `5` Luar Biasa Bagus

```python
cat_var = df.loc[:, 'Inflight wifi service':'Cleanliness']

fig, axes = plt.subplots(5, 3, figsize=(20, 20))
for cat, ax in zip(cat_var, axes.flatten()):
    sns.countplot(x=cat, data=df, hue='satisfaction', ax=ax)
```
